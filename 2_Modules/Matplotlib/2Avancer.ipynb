```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
```

On stocke dans la variable Nom_data_hydro1 le chemin du fichier à analyser et représenter 

```python
Nom_data_hydro1='./Debit_horaire.csv'
```

```python
def lecture_fichier(Nom_data_hydro1):
	donnee_date=[]
	donnee_debit_S=[]
	df = pd.read_csv(Nom_data_hydro1,delimiter=";")#, index_col=0, index='  Line'index_col='Line',, dtype={'Date':'datetime'}
	df['Date'] = pd.to_datetime(df['Date'], format='%d.%m.%Y %H:%M')#
	return (df,df['Date'],df['Debit(m3/s)'])
```

```python
[df,Da,De]=lecture_fichier(Nom_data_hydro1)
```

```python
def get_season(date):
	year = date.year
	seasons = {'Summer':(datetime.datetime(year,6,21), datetime.datetime(year,9,22)),
		'Autumn':(datetime.datetime(year,9,23), datetime.datetime(year,12,20)),
		'Spring':(datetime.datetime(year,3,21), datetime.datetime(year,6,20)),
		'winter':(datetime.datetime(year,12,21), datetime.datetime(year+1,3,20))}
	for season,(season_start, season_end) in seasons.items():
		if date>=season_start and date<= season_end:
			return season
df['Seasons']=df['Date'].apply(get_season)
```

figure 1

```python
plt.plot(Da,De, "r", linewidth=1)
plt.xlabel('Date (H)')
plt.ylabel('Debit(M3/s)')
plt.show()
```

```python
df.hist(by = 'Seasons')
```

```python
def sephist(df):
	Winter = df[df['Seasons'] == 'winter']
	Spring = df[df['Seasons'] == 'Spring']
	Autumn = df[df['Seasons'] == 'Autumn']
	Summer = df[df['Seasons'] == 'Summer']
	return Winter, Spring,Autumn ,Summer
Winter, Spring,Autumn ,Summer= sephist(df)
```

```python
plt.hist(Winter['Debit(m3/s)'], bins=[1,20,50,100,200,300], alpha=0.5, label='Winter', color='b')
plt.hist(Spring['Debit(m3/s)'], bins=[1,20,50,100,200,300], alpha=0.5, label='Spring', color='r')
plt.hist(Autumn['Debit(m3/s)'], bins=[1,20,50,100,200,300], alpha=0.5, label='Automn', color='g')
plt.hist(Summer['Debit(m3/s)'], bins=[1,20,50,100,200,300], alpha=0.5, label='Summer', color='y')
plt.legend(loc='upper right')
#n, bins, patches = plt.hist(De, bins=50, normed=1, facecolor='b', alpha=0.5)

# plt.xlabel('Mise')
# plt.ylabel(u'Probabilité')
# plt.axis([0, 150, 0, 0.02])
# plt.grid(True)
plt.show()
```

```python
name = ['Winter', 'Spring', 'Autumn', 'Summer']
data = [Winter['Debit(m3/s)'].sum(), Spring['Debit(m3/s)'].sum(),Autumn['Debit(m3/s)'].sum(),Summer['Debit(m3/s)'].sum()]

explode=(0, 0.15, 0, 0)
plt.pie(data, explode=explode, labels=name, autopct='%1.1f%%', startangle=90, shadow=True)
plt.axis('equal')
plt.show()




#p1 = plt.bar(ind, y, width, color='r') #error occurs here 
```

```python
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt

mpl.rcParams['legend.fontsize'] = 10

fig = plt.figure()
ax = fig.gca(projection='3d')
theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)
z = np.linspace(-2, 2, 100)
r = z**2 + 1
x = r * np.sin(theta)
y = r * np.cos(theta)
ax.plot(x, y, z, label='parametric curve')
ax.legend()

plt.show()
```
